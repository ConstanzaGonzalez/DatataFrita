```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.drop_duplicates("person")
mean_sku = df.groupby("person")["sku"].mean().to_frame()
mean_sku.columns = ["mean_sku"]
mean_sku
```

```python
sum_sku = df.groupby("person")["sku"].sum().to_frame()

sum_sku.columns = ["suma_sku"]
sum_sku
```

```python
min_sku = df.groupby("person")["sku"].min().to_frame()
min_sku.columns = ["min_sku"]
min_sku
```

```python
sku = mean_sku.join(min_sku)
sku = sku.join(sum_sku)
sku
```

```python
sku.to_csv('data/sku.csv')
```
