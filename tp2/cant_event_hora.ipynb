```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df["hour"] = df['timestamp'].dt.hour
```

```python
df = df[['person',"hour"]]
hours = df.groupby("person")["hour"].value_counts().to_frame()
hours.columns =["cant_event"]
hours = hours.reset_index()
hours_gt = pd.get_dummies(hours, columns = ['hour'])
```

```python
    
hours_gt = hours_gt.groupby("person").sum()
```

```python
hours_gt.to_csv('data/event_por_hora.csv',sep=",")
```
