```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df.head()
```

```python
df = df.loc[df["event"] == "checkout"]
df["model_2"] = df["model"].apply(lambda x: x.split(" ")[0])
df_m = df.groupby("person")["model_2"].value_counts().to_frame()
df_m.columns = ['cant']
df_m = df_m.reset_index()
celulares = df_m["model_2"].value_counts()
celulares
#df_m
```

```python
df_m = df_m.drop_duplicates(subset="person", keep='first')
df_m.head()
```

```python
df_m["check_apple"] = df_m["model_2"].apply(lambda x: 1 if x == "iPhone" else 0)
df_m["check_samsung"] = df_m["model_2"].apply(lambda x: 1 if x == "Samsung" else 0)
#df_m["check_motorola"] = df_m["model_2"].apply(lambda x: 1 if x == "Motorola" else 0)
#df_m["vio_mas_lg"] = df_m["model_2"].apply(lambda x: 1 if x == "LG" else 0)
df_m = df_m.set_index("person")
df_m
```

```python
df_m = df_m.drop(["model_2","cant"],1)

df_m.to_csv('data/features_checkout.csv', sep=',')
```
