```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
mezcla = df.sort_values("timestamp").drop_duplicates("person",keep="last")
mezcla.sort_values("person")

```

```python
mezcla_s = mezcla[["person","device_type","channel"]]
mezcla_s = pd.get_dummies(mezcla_s, columns=["device_type","channel"])
```

```python
mezcla_s = mezcla_s.set_index("person")
```

```python
mezcla_s.to_csv("data/channel_device.csv",sep=",")
```
