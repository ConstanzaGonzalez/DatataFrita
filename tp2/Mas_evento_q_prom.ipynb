```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

### cantidad de eventos antes de hacer una conversion

```python
df2 = df[df["event"]=="viewed product"]
df2 = df2.groupby('person')['event'].value_counts().unstack()
df2.rename(columns={"viewed product":'cant_view_prodc'}, inplace=True)
```

```python
df_prom = df2.groupby("person").agg({"cant_view_prodc":"mean"})
prom = df2.agg({"cant_view_prodc":"mean"})

df_prom["mas_prom"] = df_prom["cant_view_prodc"].apply(lambda x: 1 if x > prom.values[0] else 0)
#df_prom = df_prom.drop("cant_view_prodc",1)
```

```python
df_prom.to_csv('data/vistas_promedio.csv', sep=',')
```
