```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.drop_duplicates("person")
```

```python
test = df.loc[df["country"] != "Unknown"]
test = test.groupby("person")["country"].value_counts().to_frame()
test.columns = ['cant_de_apa']
test = test.reset_index()
test = test.drop("cant_de_apa",1)
test.info()
```

```python
test = pd.get_dummies(test,columns = ["country"])
test = test[["person","country_Brazil"]]
```

```python
test = test.reset_index()
test = test.drop_duplicates("person",keep = "first")
test = test.set_index("person")
```

```python
len(test)
```

```python
test.to_csv('data/paises_features.csv', sep=',')
```

```python
city = df
city = city.groupby("person")["region"].value_counts().to_frame()
city.columns = ['cant_de_apa']
city = city.reset_index()
city = city.drop("cant_de_apa",1)
city = city.drop_duplicates("person")
city
```

```python
city_d.to_csv('data/region_features.csv', sep=',')
```
