```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df["hour"] = df['timestamp'].dt.hour
```

```python
df = df[['person',"hour"]]
ult_hour = df.sort_values("person")
#hours = df.groupby("person")["hour"].value_counts().to_frame()
ult_hour = ult_hour.drop_duplicates("person",keep = "last")
ult_hour = ult_hour.set_index("person")
```

```python
ult_hour.to_csv("data/ult_hora.csv",sep=",")
```
