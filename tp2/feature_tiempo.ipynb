```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/events_up_to_01062018.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
features = df[["timestamp","person"]]
features = features.groupby('person')["timestamp"].value_counts().to_frame()
#features.rename(columns={evento:'cant_'+nombre}, inplace=True)
#features = features.to_frame()
features.columns = ['cant']
features.reset_index(inplace=True)
features.head(10)
```

```python

time = features.drop_duplicates(subset = ["person"],keep = "last").set_index("person").copy()
time = time.drop("cant",1)
time = time["timestamp"].astype(int).to_frame()
time.head()
```

```python
#time.to_csv('data/features_time.csv', sep=',')
```

```python
features["timestamp_int"] = features["timestamp"].astype(int).to_frame()
features_m = features.groupby("person")["timestamp_int"].mean().to_frame()
features_m.columns = ['prom_timestamp']
features_m.head()
```

```python
#features_m.to_csv('data/features_time_2.csv', sep=',')
```

```python
features_s = features.groupby("person")["timestamp_int"].sum().astype(float).to_frame()
features_s.columns = ['sum_timestamp']
features_s.head()
```

```python
features_s.to_csv('data/features_time_3.csv', sep=',')
```

```python
features = df[["timestamp","person"]]
features = features.groupby('person')["timestamp"].value_counts().to_frame()
#features.rename(columns={evento:'cant_'+nombre}, inplace=True)
#features = features.to_frame()
features.columns = ['cant']
features.reset_index(inplace=True)
features.head(10)
```

```python
time_min = features.drop_duplicates(subset = ["person"],keep = "first").set_index("person").copy()
time_min = time_min.drop("cant",1)
time_min = time_min["timestamp"].astype(int).to_frame()
time_min.columns = ['min_timestamp']
time_min.head()
```

```python
#time_min.to_csv('data/features_time_4.csv', sep=',')
```

```python
ult_view = df.loc[df["event"] == "viewed product"]
ult_view = ult_view[["person","timestamp"]]
ult_view = ult_view.groupby("person")["timestamp"].value_counts().to_frame()
ult_view.columns = ["cant"]
ult_view = ult_view.reset_index().drop_duplicates("person",keep="last").drop("cant",1)
ult_view["timestamp"] = ult_view["timestamp"].astype(int)
ult_view.columns = ["person","ult_view"]
ult_view = ult_view.set_index("person")
```

```python
ult_view.head()
```

```python
ult_check = df.loc[df["event"] == "checkout"]
ult_check = ult_check[["person","timestamp"]]
ult_check = ult_check.groupby("person")["timestamp"].value_counts().to_frame()
ult_check.columns = ["cant"]
ult_check = ult_check.reset_index().drop_duplicates("person",keep="last").drop("cant",1)
ult_check["timestamp"] = ult_check["timestamp"].astype(int)
ult_check.columns = ["person","ult_check"]
ult_check = ult_check.set_index("person")
```

```python
csv = time.join(time_min)
csv = csv.join(features_s)
csv = csv.join(features_m)
csv = csv.join(ult_view)
csv = csv.join(ult_check)
#csv["ult_view"].fillna(0, inplace=True)
csv.to_csv("data/features_time.csv",sep=",")
```
